### Mbox-short text file

```python
from google.colab import files
uploaded = files.upload()
```

```python
with open('mbox-short.txt') as f:
  context = f.read()
  print(context)
```

```python
#get total number of emails
with open('mbox-short.txt') as f:
  email_count = 0
  email_list = list()

  for line in f:
    if line.startswith('From'):
      email_count += 1
      email_list.append(line.split()[1])
  
print(email_count)
print(email_list)
```

```python
email_address = dict()

for email in email_list:
  if email in email_address:
    email_address[email] += 1
  else:
    email_address[email] = 1

print(email_address)
```

```python
unique_email_address = list(email_address.keys())
print(unique_email_address)
```
