```python
class circle():

  def __init__(self, radius):
    print('A circle with radius ', radius, ' was created.')
    self.radius = radius

  def get_circumference(self):
    return 2 * pi *  self.radius

  def get_area(self):
    return pi *  (self.radius ** 2) 

  def get_radius(self):
    return self.radius

  def __len__(self):
    return self.get_radius()

  # def __str__(self):
  #   print("This is a circle.")
  #   return "This is a circle"

  def __repr__(self):
    return "This is a circle" + str(self.get_area())
```

```python
class animal():
  
  def __init__(self, name, animal_type):
    self.name = name
    self.animal_type = animal_type
    print('Animal was created ', ' Name: ', name, ' Type : ', animal_type)

  def get_name(self):
    return 'My name :' + self.name

  def get_animal_type(self):
    return 'My type :' + self.animal_type
```

```python
animal1 = animal('timmy', 'dog')
```

```python
animal2 = animal('kitty', 'cat')
```
